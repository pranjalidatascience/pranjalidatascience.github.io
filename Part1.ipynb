```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
pop_21 = pd.read_csv('data/Canada Population 2021 Data.csv')
pop_21.head()
```

```python
pop_21_sort = pd.melt(pop_21,id_vars=['Location'],value_vars=['First_Nations','Inuit','Métis','Non_Indigenous']) 
pop_21_sort = pop_21_sort.rename(columns = {'variable':'group', 'value':'count', 'Location': 'location'})
canada = pop_21_sort[pop_21_sort['location'] == 'Canada']
canada['group'] = canada['group'].replace("Non_Indigenous", "Non Indigenous").replace("First_Nations", 'First Nations')
canada
```

```python
labels = canada['group']
colors = ['navy', 'deepskyblue', 'royalblue', 'darkgrey']
values = canada['count'].values
count = canada['count'].sum()
```

```python
fig = go.Figure(go.Pie(labels=labels, values=values, hole=.5, marker_colors = colors))

fig.update_layout(
   # title_text = "The Majority of Canadian Population is Non Indigenous",
    annotations=[
        dict(
            text=f'<span style = "font-weight:bold;font-color:grey">Canada<br>Total Population:<br>38.5 million</span>', 
            x=0.5, 
            y=0.5, 
            font_size=18, 
            showarrow=False
        )
    ], 
    height = 600, 
    width = 600, 
    showlegend = False)x
    
fig.update_traces(
    textposition='outside',
    hoverinfo='value',
    textinfo='label+percent',
    textfont_size=16,
    outsidetextfont = dict(color= ['navy', 'deepskyblue', 'royalblue', 'black']),
    hovertemplate='<span style="background-color:grey;color:black;font-family:Georgia;font-weight:bold;font-size:16px;"> %{label}<br><br>%{value:,}</span><extra></extra>', 
    hoverlabel={'bgcolor':'lightgrey'},
    marker={'line':{'color':'#000000', 'width':1}}
)

#fig.write_html('images/Donut.html')                          
fig.show()
```
