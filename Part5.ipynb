```python
pop_21 = pd.read_csv('data/Canada Population 2021 Data.csv')
pop_21.head()
pop_21['total'] = pop_21['Non_Indigenous']+pop_21['First_Nations']+pop_21['Inuit']+pop_21['Métis']
```

```python
pop_21.loc[12][1:5]
```

```python
pop_21_sort = pd.melt(pop_21,id_vars=['Location'],value_vars=['First_Nations','Inuit','Métis','Non_Indigenous']) 
pop_21_sort = pop_21_sort.rename(columns = {'variable':'group', 'value':'count', 'Location': 'location'})
canada = pop_21_sort[pop_21_sort['location'] == 'Canada']
canada['group'] = canada['group'].replace("Non_Indigenous", "Non Indigenous").replace("First_Nations", 'First Nations')
canada
```

```python
labels = ['Non Indigenous','First Nations', 'Inuit', 'Métis']
colors = ['darkgrey', 'navy', 'deepskyblue', 'royalblue']
```

```python
fig = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],horizontal_spacing=0.1,)

fig.add_trace(go.Pie(values=pop_21.loc[13][1:5],labels=labels,marker_colors = colors, hole=.4), row=1, col=1)
fig.add_trace(go.Pie(values=pop_21.loc[12][1:5],labels=labels,marker_colors = colors, hole=.4), row=1, col=2)
fig.add_trace(go.Pie(values=pop_21.loc[11][1:5],labels=labels,marker_colors = colors, hole=.4), row=1, col=3)

fig.update_layout(
   # title_text = "The Majority of Canadian Population is Non Indigenous",
    annotations=[
        dict(
            text=f'<span style = "font-weight:bold;font-color:grey">Nunavut<br>36,840</span>', 
            x=0.07, 
            y=1, 
            font_size=18, 
            showarrow=False
        )
        ,dict(
            text=f'<span style = "font-weight:bold;font-color:grey">Northwest Territories<br>41,490</span>', 
            x=0.5, 
            y=1, 
            font_size=18, 
            showarrow=False
        )
        ,dict(
            text=f'<span style = "font-weight:bold;font-color:grey">Yukon<br>43,810</span>', 
            x=0.92, 
            y=1, 
            font_size=18, 
            showarrow=False
        )
    ], 
    height = 600, 
    width = 1000, 
    showlegend = True)
    
fig.update_traces(
    textposition='auto',
    hoverinfo='value',
    textinfo='percent',
    textfont_size=16,
    outsidetextfont = dict(color= ['navy', 'deepskyblue', 'royalblue', 'black']),
    hovertemplate='<span style="background-color:grey;color:black;font-family:Georgia;font-weight:bold;font-size:16px;"> %{label}<br><br>%{value:,}</span><extra></extra>', 
    hoverlabel={'bgcolor':'lightgrey'},
    marker={'line':{'color':'#000000', 'width':1}}
)

#fig.write_html('images/Donut.html')                          
fig.show()
```
